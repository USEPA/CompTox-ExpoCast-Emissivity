# Creates the box-and-whisker and cumulative sum plots

### Load the libraries and data

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the combined average data
combined_avg = pd.read_excel(r'LR HR sum of sample concentration.xlsx', sheet_name='Combined avg')
```

```python
# Create the box and whisker plot
bplot1 = plt.boxplot(combined_avg.loc[combined_avg['Matrix'] == 'Rubber/Plastic', combined_avg.columns[-3:]], positions=(1, 1.5, 2), patch_artist=True)
bplot2 = plt.boxplot(combined_avg.loc[combined_avg['Matrix'].str.contains('Cloth'), combined_avg.columns[-2:]], positions=(3.5, 4), patch_artist=True)
bplot3 = plt.boxplot(combined_avg.loc[combined_avg['Matrix'] == 'Cloth1', combined_avg.columns[-3]], positions=(3,), patch_artist=True)
plt.xticks([1, 1.5, 2, 3, 3.5, 4], labels=['LIQ - RP', '$\mathregular{T_H}$ - RP', '$\mathregular{T_L}$ - RP', 'LIQ - CUF', '$\mathregular{T_H}$ - CUF', '$\mathregular{T_L}$ - CUF'])
colors = ['pink', 'lightblue', 'lightgreen']
bplot1['boxes'][0].set_facecolor('tab:blue')
bplot1['boxes'][1].set_facecolor('tab:orange')
bplot1['boxes'][2].set_facecolor('tab:green')
bplot2['boxes'][0].set_facecolor('tab:orange')
bplot2['boxes'][1].set_facecolor('tab:green')
bplot3['boxes'][0].set_facecolor('tab:blue')
for median in bplot1['medians']:
    median.set_color('black')
for median in bplot2['medians']:
    median.set_color('black')
for median in bplot3['medians']:
    median.set_color('black')
plt.gca().set_yscale('log')
plt.legend([bplot1["boxes"][0], bplot1["boxes"][1], bplot1["boxes"][2]], ['Liquid (LIQ)', 'High temp ($\mathregular{T_H}$)', 'Low temp ($\mathregular{T_L}$)'])
plt.ylabel('Sum of sample abundance (Î¼g/g)')
plt.tight_layout()
plt.savefig('Sum_sample_conc_box_and_whisker.png', dpi=300)
```
